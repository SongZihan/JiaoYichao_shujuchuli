```python
data = pd.read_excel('分析师预测(1).xlsx')
```

```python
for i in data['Stkcd'].unique():
    for k in ['2015-12-31','2016-12-31','2017-12-31','2018-12-31']:
        data.loc[data['Stkcd'] == i,k + "频率"]= len(data.loc[(data['Stkcd'] == i)&(data['Fenddt'] == k)])
        
```
