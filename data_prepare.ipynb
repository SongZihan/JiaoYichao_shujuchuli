```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('内部控制（一样）.xls')
# 变动项
data_b = pd.read_excel('公司治理.xlsx')

data_c = pd.read_excel('分析师预测.xlsx')

data_d = pd.read_excel('分析师实际.xlsx')
```

```python
def exce_data(a,year,i):
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['持股比例(%)']))
    except:
        a.append('')
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['其中：国有股股数']))
    except:
        a.append('')
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['董事长与总经理兼任情况']))
    except:
        a.append('')
        
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['董事人数']))
    except:
        a.append('')
        
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['其中：独立董事人数']))
    except:
        a.append('')
    try:
        a.append(float(data_b.loc[(data_b['代码'] == i)&(data_b['统计截止日期'] == year )]['监事总规模']))
    except:
        a.append('')
    return a
```

```python
def exec_data_c(a,year,i):
    try:
        a.append(data_c.loc[(data_c['Stkcd'] == i)&(data_c['Fenddt'] == year)]['Feps'].std())
    except:
        a.append('')
    try:
        a.append(data_c.loc[(data_c['Stkcd'] == i)&(data_c['Fenddt'] == year)]['Feps'].mean())
    except:
        a.append('')
    try:
        a.append(data_c.loc[(data_c['Stkcd'] == i)&(data_c['Fenddt'] == year)]['Feps'].median())
    except:
        a.append('')
    return a
```

```python
result = []
for i in data['证券代码'].unique():
    single_result = []
    # 证券代码
    single_result.append(i)
    # 证券简称
    single_result.append(data.loc[data['证券代码'] == i]['证券简称'].unique()[0])
    # 证监会行业
    single_result.append(data.loc[data['证券代码'] == i]['证监会行业'].unique()[0])
    # 辖区名称
    single_result.append(data.loc[data['证券代码'] == i]['辖区名称'].unique()[0])
    #  print(i)
    try:
        # 2019
        single_result.append(float(data.loc[(data['证券代码'] == i)&(data['报告期'] == '2018-12-31')]['内部控制指数']))
    except:
        single_result.append('')
    try:
        # 2018
        single_result.append(float(data.loc[(data['证券代码'] == i)&(data['报告期'] == '2017-12-31')]['内部控制指数']))
    except:
        single_result.append('')
    try:
        # 2017
        single_result.append(float(data.loc[(data['证券代码'] == i)&(data['报告期'] == '2016-12-31')]['内部控制指数']))
    except:
        single_result.append('')
    try:
        # 2016
        single_result.append(float(data.loc[(data['证券代码'] == i)&(data['报告期'] == '2015-12-31')]['内部控制指数']))
    except:
        single_result.append('')
    """
    到目前为止不用改变
    """
    
    single_result = exce_data(single_result,'2015-12-31',i) 
    single_result = exce_data(single_result,'2016-12-31',i) 
    single_result = exce_data(single_result,'2017-12-31',i) 
    single_result = exce_data(single_result,'2018-12-31',i) 
    """
    到目前为止不用改变
    """
    
    single_result = exec_data_c(single_result,'2015-12-31',i)
    single_result = exec_data_c(single_result,'2016-12-31',i)
    single_result = exec_data_c(single_result,'2017-12-31',i)
    single_result = exec_data_c(single_result,'2018-12-31',i)
    """
    到目前为止不用改变
    """
    
    # 实际指标
    try:
        # 2015
        single_result.append(float(data_d.loc[(data_d['Stkcd'] == i)&(data_d['Ddate'] == '2015-12-31')]['Meps']))
    except:
        single_result.append('')
    try:
        # 2015
        single_result.append(float(data_d.loc[(data_d['Stkcd'] == i)&(data_d['Ddate'] == '2016-12-31')]['Meps']))
    except:
        single_result.append('')
    try:
        # 2015
        single_result.append(float(data_d.loc[(data_d['Stkcd'] == i)&(data_d['Ddate'] == '2017-12-31')]['Meps']))
    except:
        single_result.append('')
    try:
        # 2015
        single_result.append(float(data_d.loc[(data_d['Stkcd'] == i)&(data_d['Ddate'] == '2018-12-31')]['Meps']))
    except:
        single_result.append('')
    
    result.append(single_result)
    
    
```

```python
fina = pd.DataFrame(result,columns=['证券代码','证券简称','证监会行业','辖区名称','2015年内部控制','2016年内部控制','2017年内部控制','2018年内部控制',
                                    '2015持股比例','2015年其中：国有股',
                                    '2015董事长与总经理兼任情况','2015董事人数','2015其中：独立董事人数','2015监事总规模',
                                    '2016持股比例','2016年其中：国有股',
                                    '2016董事长与总经理兼任情况','2016董事人数','2016其中：独立董事人数','2016监事总规模',
                                    '2017持股比例','2017年其中：国有股',
                                    '2017董事长与总经理兼任情况','2017董事人数','2017其中：独立董事人数','2017监事总规模',
                                    '2018持股比例','2018年其中：国有股',
                                    '2018董事长与总经理兼任情况','2018董事人数','2018其中：独立董事人数','2018监事总规模',
                                    '2015标准差','2015平均值','2015中位数',
                                    '2016标准差','2016平均值','2016中位数',
                                    '2017标准差','2017平均值','2017中位数',
                                    '2018标准差','2018平均值','2018中位数',
                                    '2015实际指标','2016实际指标','2017实际指标','2018实际指标'
                                   ])
```
